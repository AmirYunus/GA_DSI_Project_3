```python
from browsermobproxy import Server
import psutil
import time
from selenium import webdriver
import pandas as pd
```

```python
browser = webdriver.Chrome('/Users/yunora/Downloads/chromedriver')
```

```python
browser.get('https://www.alibaba.com/products/mobile_phones.html')
```

```python
next_ = browser.find_element_by_class_name('next')
```

```python
next_.click()
```

```python
products = browser.find_elements_by_xpath("//div[@class='item-main']")
```

```python
product_name = browser.find_elements_by_xpath("//div[@class='item-info']//h2[contains(@class,'title')]")
```

```python
product_price = browser.find_elements_by_xpath("//div[@class='price']")
```

```python
len(products)
```

```python
k = 1
next_ = browser.find_element_by_class_name('next')
prod_name = []
prod_price = []
while k <= 2:
    next_ = browser.find_element_by_class_name('next')
    products = browser.find_elements_by_xpath("//div[@class='item-main']")
    product_name = browser.find_elements_by_xpath("//div[@class='item-info']//h2[contains(@class,'title')]")
    product_price = browser.find_elements_by_xpath("//div[@class='price']")
    for i in range(len(products)):
        prod_name.append(product_name[i].text)
        prod_price.append(product_price[i].text)
    time.sleep(5)
    next_.click()
    k+=1
```

```python
a = pd.DataFrame(zip(prod_name,prod_price))
```

```python
a = a.rename(columns={0:'product',1:'price'})
```
